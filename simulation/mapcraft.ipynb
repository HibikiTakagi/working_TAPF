```python
con = {}
for x in range(N):
    for y in range(N):
        for z in range(N):
            con[a[x][y][z]] = []
            if y%2==0 and z+1<N:
                con[a[x][y][z]].append(a[x][y][z+1])
            
            if y%2==1 and z-1>=0:
                con[a[x][y][z]].append(a[x][y][z-1])
            
            if z%2==0 and y-1>=0:
                con[a[x][y][z]].append(a[x][y-1][z])
            
            if z%2==1 and y+1<N:
                con[a[x][y][z]].append(a[x][y+1][z])
            
            if ((y%2==0 and z%2==0) or (y%2==1 and z%2==1)) and x-1>=0:
                con[a[x][y][z]].append(a[x-1][y][z])
            if ((y%2==0 and z%2==1) or (y%2==1 and z%2==0)) and x+1<N:
                con[a[x][y][z]].append(a[x+1][y][z])

print(con)
```
