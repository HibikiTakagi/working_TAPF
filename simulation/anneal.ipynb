```python
import heapq
from collections import defaultdict

# 隣接リストを使用したグラフ構造
graph = {
    (0, 0): [(0, 1), (1, 0)],
    (0, 1): [(0, 0), (0, 2)],
    (0, 2): [(0, 1), (1, 2)],
    (1, 0): [(0, 0), (2, 0)],
    (1, 2): [(0, 2), (2, 2)],
    (2, 0): [(1, 0), (2, 1)],
    (2, 1): [(2, 0), (2, 2)],
    (2, 2): [(1, 2), (2, 1), (3, 2)],
    (3, 2): [(2, 2), (4, 2)],
    (4, 2): [(3, 2)]
}

def get_neighbors(pos, graph):
    return graph.get(pos, [])

def a_star(start, goal, graph):
    open_set = []
    heapq.heappush(open_set, (0 + manhattan_distance(start, goal), 0, start))
    came_from = {}
    g_score = defaultdict(lambda: float('inf'))
    g_score[start] = 0
    while open_set:
        _, current_cost, current = heapq.heappop(open_set)
        if current == goal:
            return reconstruct_path(came_from, current)
        for neighbor in get_neighbors(current, graph):
            tentative_g_score = g_score[current] + 1
            if tentative_g_score < g_score[neighbor]:
                came_from[neighbor] = current
                g_score[neighbor] = tentative_g_score
                f_score = tentative_g_score + manhattan_distance(neighbor, goal)
                heapq.heappush(open_set, (f_score, tentative_g_score, neighbor))
    return None

def manhattan_distance(p1, p2):
    return abs(p1[0] - p2[0]) + abs(p1[1] - p2[1])

def reconstruct_path(came_from, current):
    total_path = [current]
    while current in came_from:
        current = came_from[current]
        total_path.append(current)
    total_path.reverse()
    return total_path

class CBSNode:
    def __init__(self, paths):
        self.paths = paths
        self.cost = sum(len(path) for path in paths)

def detect_conflicts(paths):
    conflicts = []
    positions = defaultdict(set)
    for t in range(max(len(path) for path in paths)):
        pos_at_time = defaultdict(list)
        for i, path in enumerate(paths):
            pos = path[min(t, len(path) - 1)]
            pos_at_time[pos].append(i)
        for pos, agents in pos_at_time.items():
            if len(agents) > 1:
                conflicts.append((t, agents))
    return conflicts

def resolve_conflict(paths, conflict, graph, goals):
    t, agents = conflict
    constraints = []
    for agent in agents:
        current_goal_index = goals[agent].index(paths[agent][-1])
        next_goal = goals[agent][current_goal_index + 1] if current_goal_index + 1 < len(goals[agent]) else None
        if next_goal:
            new_path = a_star(paths[agent][t], next_goal, graph)
            if new_path:
                new_paths = paths[:]
                new_paths[agent] = new_paths[agent][:t] + new_path
                constraints.append(new_paths)
    return constraints

def cbs(start_positions, goals, graph):
    initial_paths = []
    for i, start in enumerate(start_positions):
        path = []
        for goal in goals[i]:
            path_segment = a_star(start, goal, graph)
            if path_segment:
                path.extend(path_segment[1:])
                start = goal
            else:
                return None
        initial_paths.append(path)
    root = CBSNode(initial_paths)
    open_set = [root]
    while open_set:
        node = heapq.heappop(open_set)
        conflicts = detect_conflicts(node.paths)
        if not conflicts:
            return node.paths
        for conflict in conflicts:
            constraints = resolve_conflict(node.paths, conflict, graph, goals)
            for constraint in constraints:
                new_node = CBSNode(constraint)
                heapq.heappush(open_set, new_node)
    return None

# テスト
start_positions = [(0, 0), (4, 2)]
goals = [[(2, 2), (4, 2)], [(2, 2), (0, 0)]]

paths = cbs(start_positions, goals, graph)
if paths:
    for i, path in enumerate(paths):
        print(f"Agent {i}: {path}")
else:
    print("No solution found")
```

```python
from heapq import heapify, heappush, heappop, heappushpop

def beam_search(root, k, api):
    """
    Args:
        root : root node
        k : number of remain paths during search
        api : apis for beam search
    Notes:
        api must have functions as follows.
        (1) init : this is called at the begenning of this function
        (2) step : return path-list or path-generator of extended path from inputed path
        (3) score : return score for path, higher scores indicate better
        (4) count : this function is called for every end of loop
        (5) terminate : return true if it should terminate to search else false
    """
    paths = [(None, root)]
    heapify(paths)
    api.init()
    while not api.terminate():
        top_paths = []
        heapify(top_paths)
        for _, path in paths:
            for extend_path in api.step(path):
                score = api.score(extend_path)
                if len(top_paths) < k:
                    heappush(top_paths, (score, extend_path))
                else:
                    heappushpop(top_paths, (score, extend_path))
        paths = top_paths
        api.count()

    result_paths = []
    result_paths_score = []
    for _, path in paths:
        result_paths.append(path)
        result_paths_score.append(score)

    return result_paths, result_paths_score
```

```python
class TSPAPI:
    def __init__(self, n):
        self.iter = 0
        self.n = n

    def init(self):
        self.iter = 0

    def step(self, path):
        for i in range(self.n):
            if i not in path:
                yield path[:] + [i]

    def score(self, path):
        """total distance of path"""
        return -sum(D[i][j] for i, j in zip(path, path[1:]))

    def count(self):
        self.iter += 1

    def terminate(self):
        return self.iter >= self.n

# TSP
N = 4
D = [
    [0, 1, 2, 3],
    [4, 0, 2, 1],
    [1, 2, 0, 3],
    [2, 3, 4, 0],
]
```

```python
root = []
k = 1
api = TSPAPI(N)
paths, scores = beam_search(root, k, api)
for path, score in zip(paths[::-1], scores[::-1]):
    print("path", path, "distance", -score)
```
